```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

First, I loaded auto-mpg.data into a pandas DataFrame and named my pandas DataFrame columns

```python
auto_header = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model', 'origin', 'car name']
auto_mpg = pd.read_fwf('https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data', header= None, names= auto_header)
auto_mpg
```

```python
auto_mpg = auto_mpg.replace(to_replace='?', value=np.nan)
auto_mpg['horsepower'] = auto_mpg['horsepower'].astype(float)
auto_mpg
```

I replaced ? in horsepower column to null values and converted horsepower column to numeric

```python
auto_mpg = auto_mpg.replace(to_replace='?', value=np.nan)
auto_mpg[['horsepower']] = auto_mpg[['horsepower']].apply(pd.to_numeric)
auto_mpg.dtypes
```

I replaced the origin column

```python
auto_mpg['origin'].replace(1, 'USA', inplace=True)
auto_mpg['origin'].replace(2, 'ASIA', inplace=True)
auto_mpg['origin'].replace(3, 'EUROPE', inplace=True)
auto_mpg
```

Created a bar chart that shows the distribution for cylinders

```python
auto_mpg.groupby("cylinders")
auto_mpg['cylinders'].plot(kind='hist', title="Distribution of Cylinders")
```

Created a scatterplot that shows the relationship between horsepower and weight.

```python
plt.scatter(auto_mpg['horsepower'], auto_mpg['weight'])
plt.suptitle('horsepower and weight')
plt.xlabel('horsepower')
plt.ylabel('weight')
```

Make conclusion on relationship between horsepower and acceleration 

```python

interesting_data = auto_mpg[['horsepower', 'acceleration',]]
relationship = sns.lmplot(x='horsepower', y='acceleration', data=interesting_data)
relationship.set_axis_labels("Horsepower", "Acceleration")
```

According to our graph, on average, cars with greater horsepower spend less time for acceleration.
